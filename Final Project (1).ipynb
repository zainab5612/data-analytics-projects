```python
1.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
2.
df = pd.read_csv("social_media_usage.csv")
df.head()
```

```python
3.
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
4. 
# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Check for duplicate rows
print("\nDuplicate rows:", df.duplicated().sum())

# Check data types
print("\nData types:\n", df.dtypes)

#remove duplicates
df = df.drop_duplicates()
```

```python
5. 
sns.histplot(df['Daily_Minutes_Spent'])
plt.title("Distribution of Daily Minutes Spent")
```

# 6. Five Questions and Answers

```python
df.groupby("App")["Daily_Minutes_Spent"].mean().sort_values()
sns.barplot(x="App", y="Daily_Minutes_Spent", data=df)
```


# 7. 
# Histogram

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Daily_Minutes_Spent"], bins=30, kde=True)
plt.title("Distribution of Daily Minutes Spent")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="App", y="Likes_Per_Day", data=df)
plt.title("Likes Per Day by App")
plt.xticks(rotation=45)
plt.show()
```

# Scatterplot

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Posts_Per_Day", y="Follows_Per_Day", data=df)
plt.title("Posts vs Follows Per Day")
plt.xlabel("Posts Per Day")
plt.ylabel("Follows Per Day")
plt.show()
```

# 8. Conclusion
In this analysis, we found that Instagram and TikTok users tend to receive more likes relative to time spent. A positive correlation exists between posting frequency and follower growth. Pinterest users spend a lot of time but gain fewer followers, showing platform differences in engagement.
